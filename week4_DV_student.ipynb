```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#%matplotlib inline
import seaborn as sns
```

# 1. Set darkgrid style from seaborn

Bar plots

```python
income = [100, 80, 150, 48, 52, 69, 88]
expense = [30, 100, 100, 20, 75, 50, 28]
index = ['A', 'B', 'C','D', 'E', 'F', 'G']
df = pd.DataFrame({'income': income,'expense': expense}, index=index)
df.head()
```

```python
sns.set_style("darkgrid")
sns.barplot(x="income",y="expense",data=df)
```

# 2. Make a bar plot

```python
fig,ax= plt.subplots()
p= np.arange(len(index))
width=0.2
ax.bar(p-width/2, income,width,label="income")
ax.bar(p+width/2, expense,width,label="expense")
```

```python
sns.set_style("darkgrid")
fig,ax= plt.subplots()
p= np.arange(len(index))
width=0.2
ax.bar(p-width/2, income,width,label="income")
ax.bar(p+width/2, expense,width,label="expense")
ax.set_xticks(p)
ax.set_xticklabels(index,rotation=90)
plt.legend()
plt.show()
```

# 5. Unstuck the bars, annotate the hight of the bars on top of them

```python
labels=["Game-1", "Game-2", "Game-3", "Game-4", "Game-5", "Game-6", "Game-7"]
sns.set_style("darkgrid")
fig,ax= plt.subplots(figsize=(9,5))
p= np.arange(len(index))
width=0.25
ax.bar(p-width/2, income,width,label="income")
ax.bar(p+width/2, expense,width,label="expence")
ax.set_xticks(p)
ax.set_xticklabels(labels)
for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x(), p.get_height()));
plt.legend()
plt.show()
```
